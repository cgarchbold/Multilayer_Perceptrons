## Loading in the data

Load and plot MNIST data


```python
# read in csvs
mnist_train1 = nn.MNIST_dataset('./data/mnist_train_0_1.csv')
mnist_train4 = nn.MNIST_dataset('./data/mnist_train_0_4.csv')

mnist_test1 = nn.MNIST_dataset('./data/mnist_test_0_1.csv')
mnist_test4 = nn.MNIST_dataset('./data/mnist_test_0_4.csv')
```

```python
# Lets quickly plot the digits, to know how to index this data
fig, axs = plt.subplots(1, 2)
fig.suptitle('Horizontally stacked subplots')


for index in range(len(mnist_test1)):
    #gather data from row
    label = mnist_test1[index]['class']
    pixels = mnist_test1[index]['image']

    #interpret one dimensional array as image
    #pixels = np.array(pixels, dtype='uint8')
    pixels = pixels.reshape((28,28))

    #set label, show image
    axs[index].set_title(f'Label is {label}')
    axs[index].imshow(pixels)#cmap ='gray')

    #stop at two examples
    if(index == 1):
        break
```
